```python
import pandas as pd
import requests

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Descargar el JSON correctamente
response = requests.get(url)
data = response.json()

# Verificar que sí se descargó bien
print(type(data))
print(len(data))

# Normalizar correctamente
df = pd.json_normalize(data)

print("Shape:", df.shape)
print("Columnas:")
print(df.columns)

df.head()
```

```python
def churn_por_categoria(columna):
    tabla = df.groupby(columna)['Churn'].mean().sort_values(ascending=False)
    return tabla
```

# 📄Informe final

## 1. Introducción

El presente análisis tiene como objetivo estudiar el fenómeno de evasión de clientes (Churn) en una empresa de telecomunicaciones.

La evasión de clientes representa uno de los principales desafíos estratégicos para empresas de servicios, ya que implica pérdida de ingresos recurrentes y mayores costos de adquisición de nuevos clientes.

A través de técnicas de limpieza de datos, análisis exploratorio y visualización, se busca identificar patrones y factores asociados al abandono del servicio, con el fin de generar insights accionables que permitan reducir la tasa de cancelación.

## 2. Limpieza y Tratamiento de Datos

Durante esta etapa se realizaron las siguientes acciones:

✅ Importación y revisión inicial

Se verificaron dimensiones del dataset.

Se identificaron tipos de datos inconsistentes.

✅ Corrección de la variable objetivo

Se detectaron valores vacíos en la columna Churn.

Se reemplazaron por valores nulos y se eliminaron los registros correspondientes.

Se convirtió la variable Churn a formato numérico (0 = permanece, 1 = cancela).

✅ Creación de nueva variable

Se creó la variable Cargo_Diario dividiendo el cargo mensual entre 30 días.

Esto permitió un análisis más granular del comportamiento de pago.

✅ Estandarización

Se homogeneizaron textos (minúsculas y eliminación de espacios).

Variables binarias fueron convertidas a formato 0/1.

Se renombraron columnas para mayor claridad.

Se aplicó codificación (One-Hot Encoding) a variables categóricas cuando fue necesario.

Estas transformaciones mejoraron la consistencia del dataset y lo prepararon para análisis estadístico.

## 3. Análisis Exploratorio de Datos (EDA)
📌 Distribución del Churn

Se analizó la proporción de clientes que permanecen vs. los que cancelan.
El dataset presenta un nivel moderado de desbalance, con mayor proporción de clientes que permanecen en el servicio.

##📌 Churn según variables categóricas

Se analizaron variables como:

Tipo de contrato

Método de pago

Servicio de internet

Género

###🔎 Hallazgos clave:

Los contratos month-to-month presentan mayor tasa de evasión.

Los clientes con métodos de pago no automáticos tienden a cancelar más.

El tipo de servicio de internet muestra diferencias relevantes en la tasa de churn.

El género no mostró una diferencia significativa.

##📌 Churn según variables numéricas

Se compararon métricas como:

Antigüedad (tenure)

Cargo mensual

Cargo total

Cargo diario

###🔎 Principales patrones:

Clientes con menor antigüedad presentan mayor probabilidad de cancelación.

Clientes con mayor cargo mensual tienden a tener mayor churn.

El gasto total es menor en clientes que cancelan, lo que sugiere abandono temprano.

## 4. Conclusiones e Insights

A partir del análisis realizado, se concluye que:

La antigüedad del cliente es un factor crítico en la evasión.

Los contratos de corto plazo (mensuales) están fuertemente asociados al churn.

El nivel de facturación mensual influye en la probabilidad de cancelación.

La estabilidad contractual y los pagos automáticos parecen reducir la evasión.

Estos hallazgos permiten identificar segmentos de alto riesgo y priorizar estrategias de retención.

## 5. Recomendaciones Estratégicas

Con base en los resultados obtenidos, se proponen las siguientes acciones:

🎯 1. Incentivar contratos de largo plazo

Ofrecer descuentos o beneficios a clientes que migren a contratos anuales o bianuales.

🎯 2. Promover métodos de pago automáticos

Implementar campañas que incentiven pagos mediante tarjeta o transferencia automática.

🎯 3. Estrategia de retención temprana

Desarrollar programas de seguimiento durante los primeros meses de servicio, ya que es el periodo de mayor riesgo.

🎯 4. Optimización de precios

Evaluar estructuras de precios para clientes con facturación alta y alto riesgo de churn.
